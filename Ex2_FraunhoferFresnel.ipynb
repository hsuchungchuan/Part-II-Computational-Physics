```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
𝜆 = 500e-9  # Wavelength of light (in meters)
𝑑 = 100e-6  # Width of the slit (in meters)
𝐷 = 1.0  # Distance from the aperture to the screen (in meters)
𝐿 = 5e-3  # Total extent of the aperture (in meters)

# Create aperture array
aperture = np.zeros(int(𝐿 / 𝜆))
start_index = int((len(aperture) - 𝑑 / 𝜆) / 2)
end_index = start_index + int(𝑑 / 𝜆)
aperture[start_index:end_index] = 1

# Perform FFT on the aperture array
aperture_fft = np.fft.fft(aperture)

# Calculate intensity pattern
intensity = np.abs(np.fft.fftshift(aperture_fft))**2

# Generate 𝑦 coordinates for the screen
𝑦 = 𝜆 * 𝐷 * np.arange(len(intensity)) / 𝐿

# Plot the intensity pattern
plt.plot(𝑦[4500:5500], intensity[4500:5500], label='Intensity Pattern')
plt.xlabel('𝑦 (m)')
plt.ylabel('Intensity')
plt.title('Diffraction Pattern')
plt.legend()

# Overlay theoretical value of the intensity pattern
theoretical_intensity = (np.sinc(𝑑 * np.sin(2 * np.pi * 𝑦 / 𝜆) / 𝜆))**2
plt.plot(𝑦[4500:5500], theoretical_intensity[4500:5500], label='Theoretical Intensity')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_diffraction_pattern(d, wavelength, D, L, num_points=1000):
    # Constants
    k = 2 * np.pi / wavelength

    # Generate y coordinates
    y_step = L / num_points
    y = np.linspace(-L/2, L/2 - y_step, num_points)  # Adjusted for proper pixel-to-distance conversion

    # Calculate diffraction pattern using FFT
    aperture = np.zeros_like(y)
    aperture[np.abs(y) <= d/2] = 1  # Single slit in the center

    # Perform FFT
    aperture_fft = np.fft.fft(aperture)
    intensity_pattern = np.abs(np.fft.fftshift(aperture_fft))**2

    # Calculate theoretical intensity pattern
    beta = np.pi * d / wavelength * np.sin(np.arctan(y / D))
    theoretical_intensity = (np.sin(beta) / beta)**2

    return y, intensity_pattern, theoretical_intensity*400

def plot_diffraction_pattern(y, intensity_pattern, theoretical_intensity):
    plt.plot(y, intensity_pattern, label='Calculated Intensity')
    plt.plot(y, theoretical_intensity, label='Theoretical Intensity', linestyle='dashed')
    plt.title('Diffraction Pattern of a Single Slit')
    plt.xlabel('Position (m)')
    plt.ylabel('Intensity')
    plt.legend()
    plt.show()

# Parameters
wavelength = 500e-9  # 500nm in meters
d = 100e-6  # 100μm in meters
D = 1.0  # Distance to screen in meters
L = 5e-3  # Aperture extent in meters

# Calculate and plot diffraction pattern
y, intensity_pattern, theoretical_intensity = calculate_diffraction_pattern(d, wavelength, D, L)
plot_diffraction_pattern(y, intensity_pattern, theoretical_intensity)
```
